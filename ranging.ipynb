```python
import pickle
```

```python
with open('bm.pickle', 'rb') as f:
    index = pickle.load(f)
```

```python
terms = list(index.keys())
```

```python
terms[:10]
```

```python
query = 'write thesauru spell'
query = set(query.split(' '))
```

```python
doc_score = {}

for t in query:
    for score, doc in index[bytes(t, 'utf-8')]:
        if doc in doc_score:
            doc_score[doc] += score
        else:
            doc_score[doc] = score
```
