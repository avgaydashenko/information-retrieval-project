```python
import os
import pickle
from tqdm import tqdm
from math import log
from nltk import PorterStemmer
import newspaper
from collections import defaultdict
```

```python
folder = 'arxiv/'
```

```python
stemmer = PorterStemmer()
url_start = '<meta name="citation_pdf_url" content="'
url_end = '" />'
authors = []
keywords = []
a_k_t_u = []
for d in tqdm(os.listdir(folder)):
    with open(folder + d) as f:
        with open(folder + d) as f1:
            htmlf=f1.read().replace('\n', '')
            for line in f.readlines():
                if url_start in line:
                    url = line[len(url_start):-len(url_end)-1]
                    article = newspaper.article.Article(url)
                    article.download(input_html=htmlf)
                    article.parse()
                    a = article.authors
                    k = article.keywords
                    t = article.title
                    authors.append(a)
                    keywords.append(k)
                    a_k_t_u.append([a, k, t, url])
```

```python
dic_author = defaultdict(list)
for aktu in a_k_t_u:
    aus = aktu[0]
    k = aktu[1]
    if len(k) == 0:
        k = ''
    t = aktu[2]
    url = aktu[3]
    for a in aus:
        dic_author[a].append(str(k) + '_' + str(t)[13:] + '_' + str(url))
```

```python
dic_author
```

```python
dic_title = {}
for aktu in a_k_t_u:
    a = aktu[0]
    k = aktu[1]
    t = aktu[2]
    url = aktu[3]
    dic_title[str(t)[13:]] = str(a)[1:-1] + '_' + str(url)
```

```python
dic_title
```

Возвращает список:
на первом месте пустой список, или список авторов и ссылка, если есть статья с таким именем
дальше возвращает списки из: ключевых слов, названия, ссылки, если упомянули в запросе автора.

```python
def check_bd(s):
    if s in dic_title:
        t = dic_title[s]
        ans = [t.split('_')]
    else:
        ans = [[]]
    terms = s.split()
    for t in terms:
        if t in dic_author:
            a = dic_author[t]
            ans.append(a[0].split('_'))
    return ans
    
```

```python
print(check_bd('Weighted Unsupervised Learning for 3D Object Detection'))
```
